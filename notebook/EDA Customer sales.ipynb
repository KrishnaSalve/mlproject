```python
fig, axs = plt.subplots(1,2, figsize = (15,7))
plt.subplot(1,2,1)
sns.histplot(df, x = 'Spending Score (1-100)', bins=30,kde=True,color='g')
plt.subplot(1,2,2)
sns.histplot(df, x = 'Spending Score (1-100)', kde=True,hue='Gender')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (15,7))
plt.subplot(1,2,1)
sns.histplot(df, x = 'Annual Income (k$)', bins=30,kde=True,color='g')
plt.subplot(1,2,2)
sns.histplot(df, x = 'Annual Income (k$)', kde=True,hue='Gender')
plt.show()
```
